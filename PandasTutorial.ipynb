**Merging DataFrames:**

---


*   Inner join
*   Outer join
*   Left join
*   Right join




```python
df1=pd.DataFrame({
    'id':[1,2,3,4],
    'name':['Rohit','Rahul','Raj','Anubhav']
})
df2=pd.DataFrame({
    'id':[3,4,5,6],
    'course':['GenAI','ML','DL','NN']
})
```

```python
# similar to intersection
inner_result=pd.merge(df1,df2,how='inner',on='id')
inner_result
```

```python
# similar to union
outer_result=pd.merge(df1,df2,how='outer',on='id')
outer_result
```

```python
left_result=pd.merge(df1,df2,how='left',on='id')
left_result
```

```python
right_result=pd.merge(df1,df2,how='right',on='id')
right_result
```

```python
# merging dataframe with different column name
df3=pd.DataFrame({
    'student_id':[1,2,3,4],
    'name':['Rohit','Rahul','Raj','Anubhav']
})
df4=pd.DataFrame({
    'id':[3,4,5,6],
    'course':['GenAI','ML','DL','NN']
})
inner_result=pd.merge(df3,df4,how='inner',left_on='student_id',right_on='id')
inner_result
```

**Concatenating DataFrames**

```python
df1=pd.DataFrame({
    'id':[1,2,3,4],
    'name':['Rohit','Rahul','Raj','Anubhav']
})
df2=pd.DataFrame({
    'id':[3,4,5,6],
    'course':['GenAI','ML','DL','NN']
})
result=pd.concat([df1,df2],axis=0, ignore_index=True)
result
```

```python
result2=pd.concat([df1,df2],axis=1)
result2
```

```python
result3=pd.concat([df1,df2],axis=0,keys=['df1','df2'])
result3
```
